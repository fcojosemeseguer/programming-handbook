# Quick Sort

```python
# Partition Function
def partition(arr, low, high):
    
    # Select the pivot
    pivot = arr[high]
    
    # Initialize the index of the smaller elements than the pivot
    i = low - 1
    
    # Tranverse from low to high looking for smaller elements than the pivot
    for j in range(low, high):
        if arr[j] < pivot:
            i+=1
            swap(arr, i, j)
        
    i += 1
    swap(arr, i, high)
    
    return i
            
            
# Swap Funciton
def swap(arr, i, j):
    arr[i], arr[j] = arr[j], arr[i]
    
    
def quickSort(arr, low, high):
    if low < high:
        
        # Pi is the index of the pivot
        pi = partition(arr, low, high)
        
        # Recurse on quickSort Function left and right side of the pivot
        
        quickSort(arr, low, pi - 1)
        quickSort(arr, pi+1, high)
     
```

```python
lst = [4, 3, 6, 1, 7, 2, 9]

quickSort(lst, 0, len(lst)-1)

print(lst)
```
