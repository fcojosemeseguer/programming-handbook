# Dynammic Programming

1- Initialize the Table

2- Fill the Table taking into account the recurring equation

3 - The lower left corner is the optim solution

(Bis) - You can rebuild the solucion backwards

# Example
### Wardrobe Fund
To celebrate your graduation, your parents give you a budget of P to spend on clothes. 
You go to a store where there are T different items of clothing, each with a price pi. Since you only wear clothes once, you want to spend as much as possible without exceeding P. 

```python
def DP(T, p, P):
    
    # 1) Initialize the Table.
    V = [ [0] * (P+1) for _ in range(T+1) ]
    
    # 2) Go through the Table, fill in it.
    for i in range(1, T+1):
        pi = p[i-1]    # current price clothe i 
        
        for e in range(1, P+1):
            
            if ((i-1) < 0)  or (e-pi < 0):
                V[i][e] = V[i-1][e]
            else:
                valor = max(V[i-1][e], pi + V[i-1][e - pi])
                V[i][e] = valor
                
    return V, V[-1][-1] # 3) The lower left corner is the optimal solution.

# 4) We can rebuild the solution backwards.
def rebuild_solution_dp(V, p):
    i = len(V) -1
    j = len(V[0]) - 1
    
    s = []
    
    
    while True:
        
        if V[i][j] == 0:
            break
        elif V[i][j] == V[i-1][j]:
            i -= 1
        else:
            s.append((i, V[i][j]))
            j -= p[i-1]
            i -=1
            
    s_fin = []
    for i in range(len(s) -1, -1, -1):
        s_fin.append(s[i])
            
    return s_fin, 
    
```

## Recurrence Relation

\[
DP[i][b] =
\begin{cases}
0 & \text{if } i = 0, e = 0  \\
-\infty\ & \text{if } i < 0, e < 0 \\
\max(DP[i-1][e],\; DP[i-1][e - p_i] + p_i) & \text{if } other case
\end{cases}
\]

## Use

```python
P = 5
p = [1, 2, 1, 2]
T = len(p)
V, optimal_value = DP(T, p, P)
path = rebuild_solution_dp(V, p)

print(optimal_value)
print(path)
V
```
